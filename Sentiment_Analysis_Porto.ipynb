# Data Cleaning & Preprocessing

```python
def normalize_unicode(text):
    return unicodedata.normalize('NFKC', text)

def cleaningText(text):
    text = normalize_unicode(str(text))
    # keep hashtag word: ubah "#kata" -> "kata"
    text = re.sub(r'#([A-Za-z0-9_]+)', r'\1', text)
    # hapus mention/RT/url/angka
    text = re.sub(r'(?<!\S)@[\w.]{1,30}\b[:;,!?]*', ' ', text)
    text = re.sub(r'\bRT\b', ' ', text)
    text = re.sub(r'http\S+', ' ', text)
    text = re.sub(r'[0-9]+', ' ', text)
    # hilangkan punctuation umum, tapi sisakan huruf & spasi
    text = text.replace('\n', ' ')
    text = text.translate(str.maketrans('', '', string.punctuation))
    # collapse whitespace
    text = re.sub(r'\s+', ' ', text).strip()
    return text

def casefoldingText(text):
    text = text.lower()
    return text

def tokenizingText(text):
    text = word_tokenize(text)
    return text

def filteringText(text):
    listStopwords = set(stopwords.words('indonesian'))
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered
    return text

def stemmingText(text):
    factory = StemmerFactory()
    stemmer = factory.create_stemmer()
    text = [stemmer.stem(word) for word in text]
    return text

def toSentence(list_words):
    sentence = ' '.join(word for word in list_words)
    return sentence
```

```python
def sentiment_analysis_lexicon_indonesia(tokens, lex_pos, lex_neg):
    score = 0
    for w in tokens:
        if w in lex_pos: score += lex_pos[w]
        if w in lex_neg: score += lex_neg[w]
    if score > 1: pol = 'positive'
    elif score < -1: pol = 'negative'
    else: pol = 'neutral'
    return score, pol
```

```python
#  pipeline per dataset
def process_dataset(csv_path, label, do_plot=False, do_lda=True, num_topics=5):
    df = pd.read_csv(csv_path)
    # clean
    df['text_clean'] = (df['full_text']
                        .apply(cleaningText)
                        .apply(casefoldingText)
                        .apply(removeSlangWords))
    df.drop(columns=['full_text'], inplace=True)
    # tokenize -> filter -> stem
    df['text_preprocessed'] = (df['text_clean']
                               .apply(tokenizingText)
                               .apply(filteringText)
                               .apply(stemmingText))
    # dedupe
    df.drop_duplicates(subset='text_clean', inplace=True)

    # sentiment
    scores_pols = df['text_preprocessed'].apply(lambda t: sentiment_analysis_lexicon_indonesia(t, LEX_POS, LEX_NEG))
    df['polarity_score'] = [sp[0] for sp in scores_pols]
    df['polarity'] = [sp[1] for sp in scores_pols]

    # save hasil antara & final
    base = f'pertamina_{label}'
    df.to_csv(f'{base}_clean_polarity.csv', index=False)

    # LDA (opsional)
    lda_info = {}
    if do_lda and len(df) > 0 and df['text_preprocessed'].str.len().sum() > 0:
        dictionary = corpora.Dictionary(df['text_preprocessed'])
        corpus = [dictionary.doc2bow(t) for t in df['text_preprocessed']]
        lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, random_state=42, passes=10)
        coherence_model = CoherenceModel(model=lda_model, texts=df['text_preprocessed'],
                                         dictionary=dictionary, coherence='c_v')
        coh = coherence_model.get_coherence()
        lda_info = {'num_topics': num_topics, 'coherence_c_v': coh}
        pd.DataFrame([{'Validation_Set': f'{base}_clean.csv',
                       'Topics': num_topics, 'Alpha': 'auto', 'Beta': 'auto', 'Coherence': coh}]
                     ).to_csv(f'{base}_lda_summary.csv', index=False)
    # Plot (opsional)
    if do_plot:
        counts = df['polarity'].value_counts()
        fig, ax = plt.subplots(figsize=(6,6))
        explode = (0.1, 0, 0) if len(counts)>=3 else None
        ax.pie(x=list(counts.values), labels=list(counts.index), autopct='%1.1f%%', explode=explode)
        ax.set_title(f'Sentiment Polarity on Tweets Data ({label})')
        plt.tight_layout()
        plt.savefig(f'{base}_polarity_pie.png', dpi=150)
        plt.close()

    return df, lda_info
```

```python
def _ensure_df(df_obj, path):
    try:
        df = df_obj.copy()
    except NameError:
        df = pd.read_csv(path)
        # pastikan text_preprocessed berbentuk list token
        if df['text_preprocessed'].dtype == 'object' and isinstance(df['text_preprocessed'].iloc[0], str):
            # jika tersimpan sebagai string daftar, coba eval aman
            df['text_preprocessed'] = df['text_preprocessed'].apply(
                lambda s: eval(s) if s.startswith('[') and s.endswith(']') else s.split()
            )
    return df

before_df = _ensure_df(globals().get('before_df', None), 'pertamina_before_clean_polarity.csv')
after_df  = _ensure_df(globals().get('after_df', None),  'pertamina_after_clean_polarity.csv')

# Standarkan urutan label sentimen agar konsisten
ORDER = ['negative', 'neutral', 'positive']
for d in (before_df, after_df):
    d['polarity'] = pd.Categorical(d['polarity'], categories=ORDER, ordered=True)
```

```python
# 2) Ringkasan metrik inti
# --------------------------------------
def sentiment_share(df):
    c = df['polarity'].value_counts().reindex(ORDER).fillna(0)
    total = c.sum() if c.sum() else 1
    pct = (c / total * 100).round(2)
    return pd.DataFrame({'count': c.astype(int), 'percent': pct})

summary_before = sentiment_share(before_df).rename_axis('sentiment').reset_index()
summary_after  = sentiment_share(after_df).rename_axis('sentiment').reset_index()

overall = pd.DataFrame({
    'metric': ['n_tweets','mean_polarity_score','median_polarity_score'],
    'before': [len(before_df), before_df['polarity_score'].mean(), before_df['polarity_score'].median()],
    'after':  [len(after_df),  after_df['polarity_score'].mean(),  after_df['polarity_score'].median()]
})
overall['delta_after_minus_before'] = overall['after'] - overall['before']
overall
```

# CNN

```python
def toSentence(tokens):
    if isinstance(tokens, list):
        return ' '.join(tokens)
    # bila tersimpan sebagai string "['a','b']"
    try:
        lst = ast.literal_eval(tokens)
        if isinstance(lst, list): return ' '.join(lst)
    except Exception:
        pass
    return str(tokens)

def ensure_df(name_or_df, path):
    """Ambil df dari memori kalau ada; kalau tidak, baca CSV."""
    if isinstance(name_or_df, pd.DataFrame):
        df = name_or_df.copy()
    else:
        df = pd.read_csv(path)
    # pastikan kolom text_preprocessed berupa list token
    if 'text_preprocessed' in df.columns and isinstance(df['text_preprocessed'].iloc[0], str):
        df['text_preprocessed'] = df['text_preprocessed'].apply(
            lambda s: ast.literal_eval(s) if s.startswith('[') and s.endswith(']') else s.split()
        )
    return df
```

```python
# LOAD HASIL PREPROCESS
# Jika before_df/after_df sudah ada di memori, fungsi akan pakai itu.
before_df = ensure_df(globals().get('before_df', None), 'pertamina_before_clean_polarity.csv')
after_df  = ensure_df(globals().get('after_df',  None),  'pertamina_after_clean_polarity.csv')
```

```python
# GABUNGKAN UNTUK TRAINING (label tetap dari kolom 'polarity')
polarity_encode = {'negative':0, 'neutral':1, 'positive':2}

df_before = before_df[['text_preprocessed','polarity']].copy()
df_before['source'] = 'before'
df_after  = after_df[['text_preprocessed','polarity']].copy()
df_after['source']  = 'after'

df_all = pd.concat([df_before, df_after], ignore_index=True)
df_all.dropna(subset=['text_preprocessed','polarity'], inplace=True)

X_text = df_all['text_preprocessed'].apply(toSentence).astype(str).values
y_raw  = df_all['polarity'].map(polarity_encode).values
```

```python
# TRAIN / TEST SPLIT (stratified)
X_train_text, X_test_text, y_train, y_test = train_test_split(
    X_text, y_raw, test_size=0.2, random_state=SEED, stratify=y_raw
)
```

```python
# HYPERPARAM
max_features = 90000   # sesuai gambar (vocab size)
embed_dim    = 200
maxlen       = 40      # panjang input sesuai gambar

# Tokenizer & padding (ulang karena max_features & maxlen berubah)
tokenizer = Tokenizer(num_words=max_features, oov_token="<OOV>", split=' ')
tokenizer.fit_on_texts(X_train_text)

X_train_seq = tokenizer.texts_to_sequences(X_train_text)
X_test_seq  = tokenizer.texts_to_sequences(X_test_text)

X_train = pad_sequences(X_train_seq, maxlen=maxlen, padding='post', truncating='post')
X_test  = pad_sequences(X_test_seq,  maxlen=maxlen, padding='post', truncating='post')

vocab_size = min(max_features, len(tokenizer.word_index) + 1)
num_classes = 3  # negative/neutral/positive
```

```python
#  ONE-HOT LABEL
num_classes = 3
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat  = to_categorical(y_test,  num_classes=num_classes)
```

```python
# CLASS WEIGHTS (imbalance-aware)
classes = np.array([0,1,2])
cw = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = {i: w for i, w in zip(classes, cw)}
```

```python
def cnn(vocab_size, maxlen=40, embed_dim=200, num_classes=3, p_drop=0.5):
    inp = Input(shape=(maxlen,))
    x = Embedding(vocab_size, embed_dim, input_length=maxlen)(inp)
    branches = []
    for k in (2,3,4,5):
        c = Conv1D(128, k, activation='relu', padding='valid')(x)
        p = GlobalMaxPooling1D()(c)
        branches.append(p)
    x = Concatenate()(branches)
    x = Dropout(p_drop)(x)
    x = Dense(256, activation='relu')(x)
    x = Dropout(p_drop/1.25)(x)
    out = Dense(num_classes, activation='softmax')(x)
    model = Model(inp, out)
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    return model

model = cnn(vocab_size=vocab_size, embed_dim=embed_dim, maxlen=maxlen)
```

```python
#  TRAINING
callbacks = [
    # EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)
]
history = model.fit(
    X_train, y_train_cat,
    validation_split=0.1,
    epochs=50,
    batch_size=256,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)
```

```python
# ---------- EVALUASI ----------
y_prob = model.predict(X_test, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['negative','neutral','positive']))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
def toSentence(tokens):
    if isinstance(tokens, list):
        return ' '.join(tokens)
    try:
        lst = ast.literal_eval(tokens)
        if isinstance(lst, list): return ' '.join(lst)
    except Exception:
        pass
    return str(tokens)

def ensure_df(obj_or_none, path):
    if isinstance(obj_or_none, pd.DataFrame):
        df = obj_or_none.copy()
    else:
        df = pd.read_csv(path)
    if 'text_preprocessed' in df.columns and isinstance(df['text_preprocessed'].iloc[0], str):
        df['text_preprocessed'] = df['text_preprocessed'].apply(
            lambda s: ast.literal_eval(s) if s.startswith('[') and s.endswith(']') else s.split()
        )
    return df
```

```python
# load hasil preprocess
before_df = ensure_df(globals().get('before_df', None), 'pertamina_before_clean_polarity.csv')
after_df  = ensure_df(globals().get('after_df',  None),  'pertamina_after_clean_polarity.csv')

polarity_encode = {'negative':0, 'neutral':1, 'positive':2}

df_before = before_df[['text_preprocessed','polarity']].copy()
df_before['source'] = 'before'
df_after  = after_df[['text_preprocessed','polarity']].copy()
df_after['source']  = 'after'

df_all = pd.concat([df_before, df_after], ignore_index=True).dropna(subset=['text_preprocessed','polarity'])
X_text = df_all['text_preprocessed'].apply(toSentence).astype(str).values
y_raw  = df_all['polarity'].map(polarity_encode).values
src    = df_all['source'].values
```

```python
# split stratified
X_train_text, X_test_text, y_train, y_test, src_train, src_test = train_test_split(
    X_text, y_raw, src, test_size=0.2, random_state=SEED, stratify=y_raw
)
```

```python
# tokenization & padding
max_features = 20000
tokenizer = Tokenizer(num_words=max_features, oov_token="<OOV>", split=' ')
tokenizer.fit_on_texts(X_train_text)

X_train_seq = tokenizer.texts_to_sequences(X_train_text)
X_test_seq  = tokenizer.texts_to_sequences(X_test_text)

lens = [len(x) for x in X_train_seq]
maxlen = max(20, int(np.percentile(lens, 95)) if lens else 50)

X_train = pad_sequences(X_train_seq, maxlen=maxlen, padding='post', truncating='post')
X_test  = pad_sequences(X_test_seq,  maxlen=maxlen, padding='post', truncating='post')

num_classes = 3
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat  = to_categorical(y_test,  num_classes=num_classes)
```

```python
# class weights
classes = np.array([0,1,2])
cw = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = {i: w for i, w in zip(classes, cw)}
```

```python
# LSTM satu arah
def build_lstm(vocab_size, embed_dim=128, lstm_units=128, dropout=0.3):
    model = Sequential([
        Embedding(input_dim=vocab_size, output_dim=embed_dim, input_length=maxlen),
        SpatialDropout1D(dropout),
        LSTM(lstm_units, return_sequences=True),   # satu arah
        GlobalMaxPooling1D(),
        Dropout(dropout),
        Dense(64, activation='relu'),
        Dropout(dropout/2),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

vocab_size = min(max_features, len(tokenizer.word_index) + 1)
lstm_uni = build_lstm(vocab_size=vocab_size, embed_dim=128, lstm_units=128, dropout=0.3)

callbacks = [
    # EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)
]

history = lstm_uni.fit(
    X_train, y_train_cat,
    validation_split=0.1,
    epochs=50,
    batch_size=256,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)
```

```python
# evaluasi
y_prob = lstm_uni.predict(X_test, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['negative','neutral','positive']))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
#  tokenization & padding
max_features = 20000
tokenizer = Tokenizer(num_words=max_features, oov_token="<OOV>", split=' ')
tokenizer.fit_on_texts(X_train_text)

X_train_seq = tokenizer.texts_to_sequences(X_train_text)
X_test_seq  = tokenizer.texts_to_sequences(X_test_text)

lens = [len(x) for x in X_train_seq]
maxlen = int(np.percentile(lens, 95)) if len(lens) > 0 else 50
maxlen = max(maxlen, 20)

X_train = pad_sequences(X_train_seq, maxlen=maxlen, padding='post', truncating='post')
X_test  = pad_sequences(X_test_seq,  maxlen=maxlen, padding='post', truncating='post')

num_classes = 3
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat  = to_categorical(y_test,  num_classes=num_classes)
```

```python
#  class weight
classes = np.array([0,1,2])
cw = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = {i: w for i, w in zip(classes, cw)}
```

```python
#  build BiLSTM
def build_bilstm(vocab_size, embed_dim=128, lstm_units=128, dropout=0.3):
    model = Sequential([
        Embedding(input_dim=vocab_size, output_dim=embed_dim, input_length=maxlen),
        SpatialDropout1D(dropout),
        Bidirectional(LSTM(lstm_units, return_sequences=True)),
        GlobalMaxPooling1D(),
        Dropout(dropout),
        Dense(64, activation='relu'),
        Dropout(dropout/2),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model
```

```python
vocab_size = min(max_features, len(tokenizer.word_index) + 1)
lstm_model = build_bilstm(vocab_size=vocab_size, embed_dim=128, lstm_units=128, dropout=0.3)

callbacks = [
    #EarlyStopping(monitor='val_accuracy', patience=10, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)
]

history_lstm = lstm_model.fit(
    X_train, y_train_cat,
    validation_split=0.1,
    epochs=50,
    batch_size=256,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)
```

```python
#  evaluasi
y_prob = lstm_model.predict(X_test, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['negative','neutral','positive']))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
# audit performa per sumber
def report_by_source(src_vals, y_true, y_hat, name):
    idx = np.where(src_vals == name)[0]
    if len(idx) == 0:
        print(f"\nTidak ada sampel {name} di test split.")
        return
    print(f"\nPerforma pada subset {name}:")
    print("Accuracy:", accuracy_score(y_true[idx], y_hat[idx]))
    print(classification_report(y_true[idx], y_hat[idx], target_names=['negative','neutral','positive']))

report_by_source(src_test, y_test, y_pred, 'before')
report_by_source(src_test, y_test, y_pred, 'after')
```

# CNN-LSTM

```python
#  Helper
def toSentence(tokens):
    if isinstance(tokens, list):
        return ' '.join(tokens)
    try:
        lst = ast.literal_eval(tokens)
        if isinstance(lst, list): return ' '.join(lst)
    except Exception:
        pass
    return str(tokens)

def ensure_df(obj_or_none, path):
    if isinstance(obj_or_none, pd.DataFrame):
        df = obj_or_none.copy()
    else:
        df = pd.read_csv(path)
    if 'text_preprocessed' in df.columns and isinstance(df['text_preprocessed'].iloc[0], str):
        df['text_preprocessed'] = df['text_preprocessed'].apply(
            lambda s: ast.literal_eval(s) if s.startswith('[') and s.endswith(']') else s.split()
        )
    return df
```

```python
#  Load data
before_df = ensure_df(globals().get('before_df', None), 'pertamina_before_clean_polarity.csv')
after_df  = ensure_df(globals().get('after_df',  None),  'pertamina_after_clean_polarity.csv')

polarity_encode = {'negative':0, 'neutral':1, 'positive':2}

df_before = before_df[['text_preprocessed','polarity']].copy()
df_before['source'] = 'before'
df_after  = after_df[['text_preprocessed','polarity']].copy()
df_after['source']  = 'after'

df_all = pd.concat([df_before, df_after], ignore_index=True).dropna(subset=['text_preprocessed','polarity'])
X_text = df_all['text_preprocessed'].apply(toSentence).astype(str).values
y_raw  = df_all['polarity'].map(polarity_encode).values
src    = df_all['source'].values
```

```python
#  Split
X_train_text, X_test_text, y_train, y_test, src_train, src_test = train_test_split(
    X_text, y_raw, src, test_size=0.2, random_state=SEED, stratify=y_raw
)
```

```python
#  Tokenizer & Padding
max_features = 20000
tokenizer = Tokenizer(num_words=max_features, oov_token="<OOV>", split=' ')
tokenizer.fit_on_texts(X_train_text)

X_train_seq = tokenizer.texts_to_sequences(X_train_text)
X_test_seq  = tokenizer.texts_to_sequences(X_test_text)

lens = [len(x) for x in X_train_seq]
maxlen = max(20, int(np.percentile(lens, 95)) if lens else 50)

X_train = pad_sequences(X_train_seq, maxlen=maxlen, padding='post', truncating='post')
X_test  = pad_sequences(X_test_seq,  maxlen=maxlen, padding='post', truncating='post')

num_classes = 3
y_train_cat = to_categorical(y_train, num_classes=num_classes)
y_test_cat  = to_categorical(y_test,  num_classes=num_classes)
```

```python
#  Class weights
classes = np.array([0,1,2])
cw = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = {i: w for i, w in zip(classes, cw)}
```

```python
#  Hybrid CNN + LSTM
def build_cnn_lstm(vocab_size, embed_dim=128, filters=64, kernel_size=5, pool_size=2, lstm_units=128, dropout=0.3):
    model = Sequential([
        Embedding(input_dim=vocab_size, output_dim=embed_dim, input_length=maxlen),
        SpatialDropout1D(dropout),
        Conv1D(filters=filters, kernel_size=kernel_size, activation='relu'),
        MaxPooling1D(pool_size=pool_size),
        LSTM(lstm_units, return_sequences=True),  # satu arah
        GlobalMaxPooling1D(),
        Dropout(dropout),
        Dense(64, activation='relu'),
        Dropout(dropout/2),
        Dense(num_classes, activation='softmax')
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])
    return model

vocab_size = min(max_features, len(tokenizer.word_index)+1)
hybrid_model = build_cnn_lstm(vocab_size=vocab_size)

callbacks = [
   # EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=2)
]

history_hybrid = hybrid_model.fit(
    X_train, y_train_cat,
    validation_split=0.1,
    epochs=50,
    batch_size=256,
    class_weight=class_weights,
    callbacks=callbacks,
    verbose=1
)
```

```python
#  Evaluasi
y_prob = hybrid_model.predict(X_test, verbose=0)
y_pred = np.argmax(y_prob, axis=1)

print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['negative','neutral','positive']))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Audit subset before/after
def report_by_source(src_vals, y_true, y_hat, name):
    idx = np.where(src_vals == name)[0]
    if len(idx) == 0:
        print(f"\nTidak ada sampel {name} di test split.")
        return
    print(f"\nPerforma pada subset {name}:")
    print("Accuracy:", accuracy_score(y_true[idx], y_hat[idx]))
    print(classification_report(y_true[idx], y_hat[idx], target_names=['negative','neutral','positive']))

report_by_source(src_test, y_test, y_pred, 'before')
report_by_source(src_test, y_test, y_pred, 'after')
```

```python
#  Plot learning curves
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,4))
ax1.plot(history_hybrid.history['accuracy'], label='train')
ax1.plot(history_hybrid.history['val_accuracy'], label='val')
ax1.set_title('Hybrid CNN + LSTM Accuracy'); ax1.legend()
ax2.plot(history_hybrid.history['loss'], label='train')
ax2.plot(history_hybrid.history['val_loss'], label='val')
ax2.set_title('Hybrid CNN + LSTM Loss'); ax2.legend()
plt.tight_layout(); plt.show()
```
